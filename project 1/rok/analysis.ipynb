```python
import pandas as pd

RESULT_DIR = "./results/"

A = pd.read_csv(RESULT_DIR + "group23.perplexityA", header=None)
B = pd.read_csv(RESULT_DIR + "group23.perplexityB", header=None)
C = pd.read_csv(RESULT_DIR + "group23.perplexityC", header=None)

A.columns, B.columns, C.columns = ["perplexityA"], ["perplexityB"], ["perplexityC"] 

print("A", A.shape)
print("B", A.shape)
print("C", A.shape)

pd.options.display.float_format = '{:.5f}'.format
results = pd.concat([A.describe(), B.describe(), C.describe()], axis=1)

A.sort_values(by=["perplexityA"], ascending=True).iloc[0:9800].hist(column="perplexityA", bins=200, figsize=(16, 7))
B.sort_values(by=["perplexityB"], ascending=True).iloc[0:9800].hist(column="perplexityB", bins=200, figsize=(16, 7))
C.sort_values(by=["perplexityC"], ascending=True).iloc[0:9800].hist(column="perplexityC", bins=200, figsize=(16, 7))

results
```
